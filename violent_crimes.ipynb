```python
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
crime_df = "crime_and_incarceration_by_state.csv"
crime_df = pd.read_csv(crime_df)
crime_df.head()
```

```python
violent_temp = crime_df.loc[crime_df['year'] == 2001].sum()
violent_total_2001 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2002].sum()
violent_total_2002 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2003].sum()
violent_total_2003 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2004].sum()
violent_total_2004 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2005].sum()
violent_total_2005 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2006].sum()
violent_total_2006 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2007].sum()
violent_total_2007 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2008].sum()
violent_total_2008 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2009].sum()
violent_total_2009 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2010].sum()
violent_total_2010 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2011].sum()
violent_total_2011 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2012].sum()
violent_total_2012 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2013].sum()
violent_total_2013 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2014].sum()
violent_total_2014 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2015].sum()
violent_total_2015 = violent_temp['violent_crime_total']
violent_temp = crime_df.loc[crime_df['year'] == 2016].sum()
violent_total_2016 = violent_temp['violent_crime_total']
```

```python
violent_summary_dict = {
    "year": [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016],
    "total": [violent_total_2001, violent_total_2002, violent_total_2003, violent_total_2004, violent_total_2005, violent_total_2006, violent_total_2007, violent_total_2008, violent_total_2009, violent_total_2010, violent_total_2011, violent_total_2012, violent_total_2013, violent_total_2014, violent_total_2015, violent_total_2016]
}
violent_summary_df = pd.DataFrame(violent_summary_dict)
violent_summary_df["total in tens of millions"] = violent_summary_df["total"]/(1000000)
violent_summary_df
```

```python
plt.plot(violent_summary_df['year'], violent_summary_df["total in tens of millions"])
plt.xlabel("Year")
plt.ylabel("Number of Violent Crimes in Tens of Millions")
plt.title("Number of Violent Crimes (2001-2016)")
plt.show()
```
